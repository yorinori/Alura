```python
confirmed_all_dates = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
confirmed_all_dates
```

O dataset utilizado durante o curso possui dados até a data '3/2/20'. Para transformar o dataset atual para que se assemelhe ao utilizado no curso podemos fazer um recorte para a data '3/2/20'. O procedimento para esta transformação é realizado a seguir.

```python
import matplotlib.pyplot as plt

""" Gráfico extremamente apertado
    estamos sendo longe do adequado ao passar uma mensagem errada
"""

plt.figure(figsize=(2,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados na China")
plt.show()
```

```python
""" Por isso em data visualization é FUNDAMENTAL
    estarmos atentos aos dois eixos
    Este gráfico está tentando comparar o crescimento do momento inicial ao momento atual
"""

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados na China")
plt.show()
```

```python
""" Exemplo Ruim de manusear o eixo y
"""

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados na China")
plt.ylim(40000, 80000)
plt.show()
```

```python
""" Exemplo Ruim de manusear o eixo y
    Aqui a mensagem ficou novamente diferente

    Este gráfico está tentando comparar o crescimento do momento inicial com 300 mil
""" 

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados na China")
plt.ylim(0, 300000)
plt.show()
```

```python
""" Exemplo Ruim de manusear o eixo y
    Aqui a mensagem ficou novamente diferente

    Este gráfico está tentando comparar o crescimento do momento inicial com 1.386 bi (população da China em 2017)

    Aqui a mensagem é novamente completamente diferente
    e pode ser passar uma mensagem ruim, até mesmo errada. poderia ser o caso de uma doença mais grave
    Ao usarmos o eixo y com a população geral, passa a ideia de que a doença não é grave

    Sendo que na realidade esse gráfico não POSSUI INFORMAÇÃO NENHUMA sobre a gravidade da doença

    Pq? Só tem caso confirmado. Não fala letalidade. Não fala incidência.
    Quais são as províncias infectadas? São de toda a China? e estou comparando com a população de toda a China.
    Errado
""" 

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Gráfico errado. Passa uma mensagem errada, com comparações injustas")
plt.ylim(0, 1386000000)
plt.show()
```

# Aula 02: Time Series, tortas e barras

```python
new_cases_china = confirmed_by_country.loc['China'][2:].diff().dropna()
```

```python
plt.figure(figsize=(10,5))
new_cases_china.plot()
plt.title("Novos casos confirmados na China")
plt.show()
```
