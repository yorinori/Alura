```python
%matplotlib inline 
import geopandas as gpd
import folium
```

```python
rj = gpd.read_file('./data/Mapas/RJ-MUNIC/RJ-MUNIC.shp')
setor= gpd.read_file('./data/Mapas/RJ-SETOR/RJ-SETOR.shp')
```

```python
crs = {'init' : 'epsg:4326'}
```

```python
rj.to_crs(crs, inplace=True)
setor.to_crs(crs, inplace=True)
```

```python
setor.fillna(0, inplace=True)
```

```python
y = rj.centroid.y.iloc[0]
y
```

```python
x = rj.centroid.x.iloc[0]
x
```

```python
data = []
for i in range(len(setor)):
    data.append([
        setor['geometry'][i].centroid.y,
        setor['geometry'][i].centroid.x,
        setor['V002'][i]/setor['V002'].max()
    ])
```

```python
data
```

```python
from folium.plugins import HeatMap
```

```python
base = folium.Map(
    [y, x],
    zoom_start=11,
    tiles='OpenStreetMap'
)

HeatMap(data, name="População").add_to(base)
folium.LayerControl().add_to(base)
base.save('HeatMap.html')

base
```
